```python
continent_cases = data.groupby("continent")["Cases"].sum()

print(continent_cases)
```

```python
import matplotlib.pyplot as plt

continent_cases = data.groupby("continent")["Cases"].sum()

plt.figure(figsize=(8,5))

plt.plot(
    continent_cases.index,
    continent_cases.values,
    marker='o',
    linestyle='-'
)

plt.title("Global COVID-19 Cases by Continent")
plt.xlabel("Continent")
plt.ylabel("Total Cases")

plt.grid(True)
plt.xticks(rotation=45)

plt.show()
```

```python
continent_deaths = data.groupby("continent")["Deaths"].sum()

plt.figure(figsize=(8,5))

plt.plot(
    continent_deaths.index,
    continent_deaths.values,
    marker='o',
    linestyle='-'
)

plt.title("Continent-wise COVID-19 Deaths Trend")
plt.xlabel("Continent")
plt.ylabel("Total Deaths")

plt.grid(True)
plt.show()
```

```python
numeric_data = data.select_dtypes(include=['number'])

corr = numeric_data.corr()

plt.figure(figsize=(10,6))

sns.heatmap(
    corr,
    annot=True,
    cmap="coolwarm",
    linewidths=0.5
)

plt.title("Correlation Heatmap of COVID-19 Dataset")
plt.show()
```

Insights

1.Cases and deaths show a strong positive correlation, meaning higher infections generally lead to more deaths.

2.Europe, Asia, and North America are the most affected continents in terms of total COVID-19 impact.

3.Africa and Oceania show comparatively lower cases and deaths than other continents.

4.Recovery patterns indicate that most infected people eventually recovered, showing improvement in healthcare response.

5.Data suggests COVID-19 impact varies widely by region due to population, healthcare, and reporting differences.
